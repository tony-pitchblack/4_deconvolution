```python
size = 15
sigma = 5
center = (size - 1) / 2
x = np.arange(size)
y = np.arange(size)
xv, yv = np.meshgrid(x, y)
rvar = np.exp(-((xv - center)**2 + (yv - center)**2) / (2 * sigma**2)) / (2 * sigma**2)
plt.imshow(rvar, cmap='gray')
plt.title(f"Gaussian noise (size={size}, sigma={sigma})")
plt.colorbar()
plt.axis('off')
plt.show()
```

```python
import numpy as np
from scipy.signal import convolve2d

def gaussian_kernel(size, sigma):
    center = (size - 1) / 2
    x = np.arange(size)
    y = np.arange(size)
    xv, yv = np.meshgrid(x, y)
    rvar = np.exp(-((xv - center)**2 + (yv - center)**2) / (2 * sigma**2)) / (2 * sigma**2)
    return rvar
```
